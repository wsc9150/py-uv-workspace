# 모듈과 패키지
모듈
- 모듈은 서로 연관된 함수, 변수,  클래스들을 모아놓은  ‘단일 파이썬 파일 (.py)’

특징
- 코드의 재사용성을 높이고 관리를 쉽게 하기 위해 파일을 분리한 가장 작은 단위
- import 키워드를 통해 다른 파일에서 불러와 사용 가능
 
모듈의 종류  
- 표준 모듈 : 파이썬에 기본으로 설치된 모듈과 패키지, 내장 함수를 묶어서 파이썬 표준 라이브러리(Python Standard Library, PSL)라 부른다.
https://docs.python.org/ko/3/library/index.html

-  사용자 생성 모듈 : 사용자가 만드는 모듈
- 서드파티(Third Party) : 협력 업체나 개인이 만들어서 제공하는 모듈

````python
%%writefile module_mymath_01.py

mypi = 3.14

def add(a, b):
    return a + b

def area(r):
    return mypi * r * r
````

## 모듈 가져오기(import)  
- 다른 프로그램으로부터 데이터를  갖고 오는 것을 의미한다.  
- import 방법

| 형식 | 의미 | 예제 |
| :--- | :--- | :--- |
| `import 모듈` | 모듈 전체를 가져오는 형식 | `import math` |
| `from 모듈 import *` | 모듈 전체를 가져오는 형식 | `from math import *` |
| `from 모듈 import 변수 or 함수 or 클래스` | 모듈 내에서 필요한 것만 가져오는 형식 | `from math import sin, cos, pi` |
| `import 모듈 as 별칭` | 모듈을 다른 이름으로 가져오는 형식 | `import numpy as np` |
| `from 모듈 import 변수 or 함수 or 클래스 as 별칭` | 모듈 내의 특정 요소를 다른 이름으로 가져오는 형식 | `from re import sub as substitute` |
| `from 모듈 import 변수 or 함수 or 클래스 as 별칭1, 변수 or 함수 or 클래스 as 별칭2` | 모듈 내의 여러 요소를 각각 다른 이름으로 가져오는 형식 | `from re import sub as sub1, subn as sub2` |
| `from 모듈 import (변수 or 함수 or 클래스, ...)` | 하나의 모듈에서 여러 개의 이름을 가져올 때 괄호를 사용하는 형식 | `from Tkinter import (Tk, Frame, Button, Entry)` |

````python
# import 모듈명 방식 : 모듈 이름을 접두사로 붙여 변수와 함수에 명확하게 접근하는 표준 방식
import module_mymath_01
print(dir(module_mymath_01))
print(module_mymath_01.mypi)
print(module_mymath_01.area(5))
````

````python
from module_mymath_01 import mypi, add, area
print(mypi)
print(add(10, 20))
print(area(5))
````

````python
# import 모듈명 as 별명 방식 : 모듈 이름이 너무 길거나 복잡할 때 축약형 별칭(Alias)을 부여하는 방식
import module_mymath_01 as mm
print(dir(mm))
print(mm.mypi)
print(mm.add(10, 20))
print(mm.area(5))
````

## 6.2 패키지(package)
### 패키지(package)

- 여러 개의 파이썬 모듈(.py)을 하나의 폴더 구조로 묶어 관리하는 단위
- 점(.) 연산자를 사용해 계층적으로 모듈을 분리하여 대규모 프로젝트 관리에 필수
- VS Code 실습 프로젝트 최종 폴더 트리 구조
```
  my_workspace/             # 1. VS Code로 열어둔 작업 상위 폴더
  ├── main.py               # 2. 패키지를 불러와 실행할 스크립트 파일
  └── my_shapes/            # 3. 개발자가 직접 만든 '커스텀 패키지' 폴더
      ├── __init__.py       #    - 이 폴더가 패키지임을 정의하는 파일 (비어있어도 됨)
      ├── area.py           #    - 면적 계산 모듈
      └── volume.py         #    - 부피 계산 모듈
```

```
■ 작업 순서
  ① VS Code 좌측 탐색기(Explorer) 빈 곳 우클릭 ──> [New Folder] 선택 ──> 폴더명 'my_shapes' 입력
  ② 생성된 'my_shapes' 폴더 안에 새 파일 [New File] 생성 ──> 파일명 '__init__.py' 입력 (내용은 빈 상태로 유지)
  ③ 같은 폴더 안에 새 파일 [New File] 생성 ──> 파일명 'area.py' 입력 후 아래 코드 작성

■ [my_shapes/area.py] 소스 코드
  def circle_area(r):
      return 3.14 * r * r


  def square_area(w, h):
      return w * h
```

````python
%mkdir my_shapes
````

```
__init__.py 파일의 실무적 역할
  - 파이썬 3.3 버전부터는 이 파일이 없어도 패키지로 인식되지만, 구버전 호환성과 패키지 초기화 코드를 내장하기 위해 빈 파일이라도 생성해 주는 것이 글로벌 표준 스펙이다.
```

````python
%%writefile my_shapes/area.py

def circle_area(r):
    return 3.14 * r * r


def square_area(w, h):
    return w * h
````

### Step 2. 외부에서 패키지 모듈 호출 및 실행하기
```
■ 작업 순서
  1. 'my_shapes' 폴더 밖(상위 폴더 위치)에 새 파일 [New File] 생성 ──> 파일명 'main2.py' 입력
  2. 폴더 계층 구조를 참조하여 패키지 내 함수를 호출하는 코드 작성 후 저장


■ [main2.py] 소스 코드
  # from 패키지명.모듈명 import 함수명 구조로 명확하게 호출
  from my_shapes.area import circle_area, square_area


  print("■ 커스텀 패키지 실행 결과 ■")
  print(f"반지름 5인 원의 넓이 : {circle_area(5)}")         # 출력: 78.5
  print(f"가로4 세로5 사각형 넓이 : {square_area(4, 5)}")    # 출력: 20
```


````python
%%writefile main2.py

from my_shapes.area import circle_area, square_area

print("■ 커스텀 패키지 실행 결과 ■")
print(f"반지름 5인 원의 넓이 : {circle_area(5)}") # 출력: 78.5
print(f"가로4 세로5 사각형 넓이 : {square_area(4, 5)}") # 출력: 20
````

````python
%%writefile main2.py

import my_shapes.area as ma

print("■ 커스텀 패키지 실행 결과 ■")
print(f"반지름 5인 원의 넓이 : {ma.circle_area(5)}") # 출력: 78.5
print(f"가로4 세로5 사각형 넓이 : {ma.square_area(4, 5)}") # 출력: 20
````

````python
!python main2.py
````
