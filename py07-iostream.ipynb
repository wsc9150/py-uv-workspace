## 파일 입출력

- 파이썬 표준 함수(내장함수)에는 파일과 관련된 처리를 하는 함수가 기본으로 제공된다.
- 파일은 텍스트 파일과 바이너리 파일로 나뉜다.
- 파일을 열때 open( )함수를 사용한다.
- 파일 입출력 순서
 ① 어플리케이션이 운영체제에게 파일 처리를 API 함수를 통해 의뢰하면, 운영체제가
   요청한 업무를 수행해 주고 그 결과를 어플리케이션에게 돌려줌.
 ② 어플리케이션이 파일 처리 업무를 의뢰하는 과정과 각 과정에서 사용되는 파이썬
    함수






| 단계 | 처리 흐름 | 매칭 소스 코드 | 세부 설명 |
|---|---|---|---|
| ① | 파일 열기 | file = open() | open() 내장함수로 파일을 연다. |
| ② | 파일 읽기/쓰기 | file.read() / file.write() | 얻어진 파일 객체에서 데이터를 읽고 쓴다.(read,write) |
| ③ | 파일 닫기 | file.close() | close() 함수로 객체의 사용을 종료한다. |




파일객체 = open('파일경로', '파일열기모드')

````python
fread = open('source/sample.txt', 'r')

while True:
    line = fread.readline() # readline()은 파이레서 한 줄씩 읽어옴
    if not line:            # 파일에서 더 이상 읽어올 내용이 없으면
        break               # 현재 수행중인 반복문을 빠져나온다.

    print(line, end='')

fread.close()               # 파일 닫기
````

````python
fread = open('source/sample.txt', 'r')
line = fread.readlines() # readlines() 함수는 파일에서 한 라인씩 읽어와서 list에 저장한 후 return
print(line)
print(type(line))
fread.close()

for data in line:
    print(data, end='')
````

## 주요 입출력(직렬화/데이터 저장) 모듈
직렬화(Serialization)란?  
- 파이썬의 객체(리스트, 딕셔너리, 클래스 등)를 파일로 저장하거나 네트워크로 전송할 수 있는 바이트/텍스트 형태로 변환하는 작업입니다.
- 입출력 모듈 종류  
① pickle: 텍스트 상태의 데이터가 아닌 파이썬 객체 자체를 파일로 저장하는 모듈  
② json : 타 언어 호환성이 가장 뛰어난 텍스트 기반 모듈    
③ csv : 엑셀과 연동되는 표 형태 데이터 전용 모듈  
④ marshal : 파이썬 내부 바이트코드를 위한 고속 모듈    
⑤ shelve : 파일 자체를 딕셔너리처럼 쓰는 영구 저장소 모듈  
⑥ configparser : 설정 파일(INI)을 관리하는 전용 모듈    

| 모듈명 | 데이터 형태 | 타 언어 호환 | 보안성 | 주요 용도 |
| :---: | :---: | :---: | :---: | :--- |
| **pickle** | 바이너리 | 불가능 (파이썬 전용) | ⚠️ 취약 (경고) | 파이썬 복잡한 객체 완벽 저장 |
| **json** | 텍스트 (문자열) | **우수 (글로벌 표준)** | 안전함 | 웹 API 통신, 일반 데이터 저장 |
| **shelve** | 바이너리 | 불가능 (파이썬 전용) | ⚠️ 취약 (경고) | 대용량 key-value 로컬 데이터베이스 |
| **csv** | 텍스트 (문자열) | **우수 (엑셀 연동)** | 안전함 | 표 형태 데이터, 데이터베이스 내보내기 |


### pickle 모듈이란?  
- 파이썬 객체 구조를 그대로 바이트 스트림으로 변환(직렬화)하고, 이를 다시 파이썬 객체로 복원(역직렬화)하는 파이썬 고유의 표준 모듈

- 핵심 함수  
① pickle.dump(obj, file): 객체를 바이너리 파일에 저장  
② pickle.load(file): 바이너리 파일로 부터 객체를 복원

- ⚠치명적인 보안 경고 (중요)  
① pickle은 신뢰할 수 없는 데이터(인터넷에서 다운로드한 파일 등)를 역직렬화할 때    악성 코드가 자동으로 실행될 수 있는 심각한 보안 취약점이 있다.  
② 따라서 외부 통신용 데이터 저장에는 절대로 사용을 지양해야 한다.

````python
import pickle

with open('source/sample.txt', 'wb') as f:
    pickle.dump(['apple', 'banana', 'lemon'], f) # dump(저장할 데이터, 파일)
````

````python
with open('source/sample.txt', 'rb') as f:
    print(pickle.load(f))
````

### json (웹 표준 및 안전한 대체재)
- json 모듈?
  1. 텍스트 기반(문자열)으로 저장되어 사람이 직접 읽고 수정할 수 있음
  2. 자바스크립트, 자바, C++ 등 다른 프로그래밍 언어와 데이터 교환 시 표준으로 사용
  3. 텍스트 형식이므로 pickle과 같은 악성 코드 실행 위험이 없어 매우 안전함
- 핵심 암기 공식: 's' = String (문자열)
  - s가 없는 함수 (dump, load)   →  '파일(File)' 객체를 직접 읽고 쓸 때 사용
  - s가 있는 함수 (dumps, loads) →  '문자열(String)' 데이터를 직접 다룰 때 사용


- 4대 함수 기능 요약
  1. json.dump()  : 파이썬 객체를 'JSON 파일'로 저장 (입력)
  2. json.dumps() : 파이썬 객체를 'JSON 문자열'로 변환 (변환)
  3. json.load()  : 'JSON 파일'을 파이썬 객체로 읽어옴 (출력)
  4. json.loads() : 'JSON 문자열'을 파이썬 객체로 변환 (복원)


- json.dump() / dumps() 핵심 옵션
  - 파이썬 객체를 JSON 형식으로 변환할 때 가독성, 정렬, 한글 처리, 데이터 예외 등을 세밀하게 제어하기 위해 인자(Argument) 값을 추가할 수 있다.
  1. indent : 들여쓰기 공백 수 지정 (가독성 향상)
  2. ensure_ascii : 아스키(ASCII) 변환 여부 (한글 깨짐 방지)
  3. sort_keys : 딕셔너리 키(Key) 정렬 여부 (데이터 정렬)
  4. separators : 데이터 구분 기호 변경 (용량 압축)
  5. default : 직렬화 불가능한 객체의 처리 함수 지정 (에러 방지)

***문자열 전용 함수 (dumps / loads) 예제***
- 파일 입출력(I/O) 과정 없이, 메모리 상에서 파일 형식을 가진 문자열로만 주고받을 때 사용합니다. (웹 통신, API 전송 등)

````python
import json

data = { 'name': 'kim', 'age': 25 }
json_string = json.dumps(data)  # dict -> str
print(json_string)              # '{"name": "kim", "age": 25}'
print(type(json_string))        # str
````

````python
json_dict = json.loads(json_string)
print(json_dict)
print(type(json_dict))
````

````python
import json
from datetime import datetime
    
data = {
    'c_group': '인공지능팀',
    'a_id': 1004,
    'd_data': [1, 2, 3],
    'b_date': datetime.now()
}


# datetime 객체를 문자열로 변환
def json_default_handler(obj):
    if isinstance(obj, datetime):
        return obj.strftime('%Y-%m-%d %H:%M:%S')

result = json.dumps(
    data,
    indent=4,
    ensure_ascii=False,
    sort_keys=True,
    default=json_default_handler # 기본 직렬화할 수 없는 객체를 만났을 때 호출할 함수 지정
)

print(result)
print(type(result))
````

```
  ※ 특이사항 분석

  - 한글 '인공지능팀'이 깨지지 않고 정상 표기됩니다.
  - 변환 불가였던 datetime 객체가 지정한 문자열 포맷으로 안전하게 출력되었습니다.
```

````python
header = ['학번', '이름', '전공']
students = [
    ['202601', '홍길동', '컴퓨터공학'],
    ['202602', '이순신', '데이터사이언스'],
    ['202603', '유관순', '인공지능학'],
]

# 1 : 반복문
student_list = []
for student in students:
    student_list.append(dict(zip(header, student)))

# 2 : list comprehension
student_list = [ dict(zip(header, student)) for student in students ]

# 3 : lambda
student_list = list(map(lambda x: dict(zip(header, x)), students))

# print(student_list)

result = json.dumps(student_list, indent=4, ensure_ascii=False)
print(result)
````

````python
with open('source/students.json', 'w') as f:
    json.dump(student_list, f, ensure_ascii=False, sort_keys=True)
````

````python
with open('source/students.json', 'r') as f:
    loaded_data = json.load(f)

print(loaded_data)
print(type(loaded_data))
````

### csv (Comma-Separated Values)
- 표 형태의 데이터를 쉼표(,)로 구분하여 저장하는 표준 모듈
- 엑셀(Excel)이나 데이터 분석 라이브러리(Pandas)와 완벽하게 연동됨
- newline='' 옵션: 윈도우 환경 등에서 데이터 저장 시 줄바꿈이 중복되는 현상을 방지
- encoding='utf-8-sig': 한글 데이터가 엑셀(Excel)에서 깨지지 않도록 인코딩 적용

````python
import csv

header = ['학번', '이름', '전공']
students = [
    ['202601', '홍길동', '컴퓨터공학'],
    ['202602', '이순신', '데이터사이언스'],
    ['202603', '유관순', '인공지능학'],
]

# csv로 데이터 저장
with open('source/records.csv', 'w', newline='') as f: # newline='' : 저장할 때 새로운 라인이 추가되지 않도록 설정
    writer = csv.writer(f)
    writer.writerow(header)     # 헤더(제목)
    writer.writerows(students)  # 전체 데이터

# csv 파일 내용 읽어오기
with open('source/records.csv', 'r') as f:
    reader = csv.reader(f)
    print(reader)

    for row in reader:
        # print(row)
        print(f'학번 : {row[0]} | 이름 : {row[1]} | 전공 : {row[2]}')
````
