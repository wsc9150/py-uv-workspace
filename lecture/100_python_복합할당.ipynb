### starred
unpacking할당할때, *를 붙이면 좌우변의 개수가 달라도 에러가 발생하지 않는다.

```python
a, *b = 1,2,3,4  # *를 붙여기 때문에 Error 발생하지 않음
```

```python
a
```

```python
b
```

*는 나머지 의미이기 때문에 단독으로 사용할 수 없다.

```python
*b = 1,2,3,4
```

콤마(,)를  붙여주면 전체를 나머지로 받아올 수 있다.  
이때는 다른 이름이 있을 수도 있으므로 나머지 개념이 성립한다.

```python
*b, = 1,2,3,4
```

```python
b
```

*2개를 사용할 수 없다.

```python
*b,*c = 1,2,3,4
```

```python
*b,*c, = 1,2,3,4
```

### 조건/판별

논리연산자 (and/or)

```python
a = 1
```

```python
a > 2
```

```python
a < 4
```

and는 앞에서 True이면 뒤에 것을 반환하고,  
앞에서 False이면 앞에 것을 반환한다.

```python
a > 2 and a < 4
```

or는 앞에서 True이면 앞에 것을 반환하고,   
앞에서 False이면 뒤에 것을 반환한다.

```python
a > 2 or a < 4
```


interning

파이썬은 느리기 때문에 중간에 cache 저장하는 기법이 있다.  
-5에서 256은 파이썬에서 같은 메모리 주소를 사용한다. (interning 기법)

```python
a = 100
b = 100
```

```python
# == 메모리에 저장된 값을 비교
a == b
```

```python
# is은 메모리 주소를 비교
a is b
```

```python
a = 257
b = 257
```

```python
a == b
```

```python
a is b
```

```python
a = 5
b = 8
```

```python
a == b
```

```python
a is b
```

파이썬에서 문자열 일때는 a-zA-Z0-9_ 는 내부적으로 interning 기법을 사용한다.

```python
a = 'python'
b = 'python'
```

```python
a == b
```

```python
a is b
```

```python
a = 'python3'
b = 'python3'
```

```python
a == b
```

```python
a is b
```

```python
a = '파이썬'
b = '파이썬'
```

```python
a == b
```

```python
# 한글은 무조건 다른 메모리를 할당함
a is b
```

```python
a = 'python version'
b = 'python version'
```

```python
a == b
```

```python
# 영문에 공백이 포함되여 있으면 다른 메모리에 할당이 됨
a is b
```

영문자일때 공백포함되여 있어도 interning기법을 사용할 수 있도록 intern을 사용하면 된다.

```python
from sys import intern
```

```python
c = intern('python programming')
d = intern('python programming')
```

```python
c == d
```

```python
c is d
```

```python
# 숫자일때는 intern()을 사용할 수 없다.
a = intern(255)
b = intern(255)
```
