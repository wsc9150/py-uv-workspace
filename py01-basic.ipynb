# 자료형


1. 데이터 저장방법에 따른 자료형의 분류  
  직접(direct)형 : 직접 데이터를 표현하는 자료형 : int, float, complex      
  시퀀스(sequence)형: 다른 데이터를 포함하는 자료형, 순서가 있는 집합 자료형 : list,str, tuple, bytes, bytearray,range    
  매핑(Mapping)형 : 다른 데이터를 포함하는 자료형 : dict    
  집합(Set)형 : 순서가 없고, 중복된 값도 없다. : set, frozenset  


2. 변경 가능성에 따른 자료형 분류  
  변경 가능형(Mutable) : 데이터의 값을 변경할 수 있다.  : list, dict, set  
  변경 불가능형(Immutable) : 데이터의 값을 변경할 수 없다. : int, float, complex, str, tuple, frozenset


3. 데이터 저장 개수에 따른 자료형의 분류  
  리터널(Literal/Scalar)형 : 한 가지 객체만 저장한다. : int, bytes, str, float, complex  
  컨테이너(Container)형 : 여러 가지 객체를 저장한다. : list, tuple, dict, set, fronzenset

문자열 포매팅(서식지정) : 어떤 문서 양식을 만들어 넣고 빈칸에 필요한 내용을 채워서 문서를 지정한다.


문자열포매팅(서식지정) 형식


|코드|설명 |
|---|:---:|
|%d|정수|
|%f|실수|
|%s|문자열|
|%c|하나의문자|
|%o|8진수|
|%x|16진수|
|%%|%기호자체를출력하고자할때|



```python
s_val7 = 'python'
s_val8 = 83.5

print('%s 의 점수는 %.1f입니다.' % (s_val7, s_val8)) # python1, 2방식
print('{} 의 점수는 {}입니다.'.format(s_val7, s_val8)) # python3 방식
print('{1} 의 점수는 {0}입니다.'.format(s_val8, s_val7))
print('{subject} 의 점수는 {jumsu}입니다.'.format(jumsu=s_val8, subject=s_val7))
print('{{0}} 의 점수는 {{1}}입니다.'.format(s_val7, s_val8)) # {}자체를 출력할때 {{}}을 한다.
print('{{{0}}} 의 점수는 {{{1}}}입니다.'.format(s_val7, s_val8)) # {}자체와 값을 출력할때는 {{{}}}을 한다.

print('{0:<10} 의 점수는 {1}입니다.'.format(s_val7, s_val8)) # 10자리 확보하고 남은 공간은 오른쪽을 여백으로 채운다.
# print('{:<10} 의 점수는 {}입니다.'.format(s_val7, s_val8))

print('{0:>10} 의 점수는 {1}입니다.'.format(s_val7, s_val8))
```

집합(set)
- 집합은 여러 값을 순서 없고 중복 없이 모아놓은 자료형
- set, frozenset 두 종류
- set은 변경이 가능한 집합이고 frozenset은 변경이 불가능한 자료형
- 중복 제거 및 교집합, 차집합, 합집합 등 수학연산 수행에 사용
- 순서 유지가 안되기 때문에 indexing 및 slicing 사용 불가능
- 반복 가능한 객체로 집합 객체를 생성할 수 있다.

```python
s = set()
print(s)
print(type(s))
print(dir(s))
```

```python
# set(iterable)
print(set((3, 1, 5, 2, 5)))

a = (3, 1, 5, 2, 5)
b = set(a)
print(b)
```

```python
print(set('apple'))
print(set([3, 1, 5, 2, 5]))
```

```python
a = [3, 1, 5, 2, 5]
b = set(a) # deep copy
print(b)

a[0] = 6
print(a)
print(b)
```

```python
k = [2, 4]
a = [3, 1, k, 5]
print(a)

b = set(a)
print(b)
```

```python
# dict 자료형으로 set을 생성하면 키 값으로만 생성
b = set({ 'one': 1, 'two': 2 })
print(b)
print(type(b))
```

```python
a = set((1,2,3,4,5))
b = set((4,5,6,7))
print(a.union(b))           # 합집합
print(a | b)                # 합집합

print(a.intersection(b))    # 교집합
print(a & b)                # 교집합

print(a.difference(b))      # 차집합
print(a - b)                # 차집합
```

```python
# 요소 추가
a = set((1, 2, 3))
a.add(4)
print(a)
```

```python
# 요소 여러개 추가
a = set((1, 2, 3))
a.update((4, 5))
print(a)
```

```python
# 요소 제거
a = set((1, 2, 3))
a.remove(2)
print(a)
```

dict
- 특정 키를 주면 이와 관련된 값을 돌려주는 내용 기반으로 검색하는 자료형
- 데이터의 저장 순서가 없다.
- 내부적으로 해시 기법을 사용하기 때문에 검색 속도가 빠르다.
- key-value 형식으로 저장
- key 값은 중복을 허용하지 않는다.

```python
d = {}
print(d)
print(type(d))
print(dir(d))
```

```python
# key 값을 통한 value 접근
d = { 'one': 1, 'two': 2 }
print(d)
print(d['one'])
```

```python
d = dict(one=1, two=2)
print(d)
print(type(d))
print(d['one'])
```

```python
# zip 함수를 이용하여 생성
keys = ['one', 'two', 'three']
values = [1, 2, 3]

a = zip(keys, values)
print(a)

d = dict(a)
print(d)
print(d['one'])
```

```python
keys=['one', 'two', 'three', 'four', 'five']
values=[1, 2, 3]

# keys, values의 크기가 다르면 작은수에 맞춰 생성된다.
a = zip(keys, values)
print(a)

d = dict(a)
print(d)
print(d['one'])
```

```python
d = { 'one': 1, 'two': 2 }
print(d.keys())

for k in d.keys():
    print(k)

print(d.values())

for k in d.values():
    print(k)
```

```python
d = { 'one': 1, 'two': 2}
print(d['one'])
print(d.get('one'))
print(d.get('three'))       # key 값이 없으면 None 리턴
print(d.get('three', 0))    # get(key, defualt)
```

```python
# key-value 삽입
d = { 'one': 1 }
d['two'] = 2
print(d)
```

```python
# dict 요소 제거
d = { 'one': 1, 'two': 2 }
print(d.pop('one')) # 요소를 꺼내면서 제거
print(d)

d = { 'one': 1, 'two': 2 }
d.pop('one') # 요소 제거
print(d)
```

```python
# 전체 제거
d = {'one': 1, 'two': 2}
d.clear()
print(d)
```
