[문제1]  
문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 정의하세요.    
make_url("naver")  

[출력결과]    
www.naver.com  

```python
def make_url(url):
    newUrl = 'www.' + url + '.com'
    return newUrl

print(make_url('naver'))
```

[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pickup_even(numList):
    newList = list(filter(lambda x: x % 2 == 0, numList))
    return newList

print(pickup_even([3, 4, 5, 6, 7, 8]))
```

[문제4]  
콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하세요.

convert_int("1,234,567")

[출력결과]   
1234567

```python
def convert_int(num):
    newNumber = int(num.replace(',', ''))
    return newNumber

print(convert_int('1,234,567'))
```

[문제4]  
세 개의 숫자를 입력받아 가장 큰수를 출력하는 print_max 함수를 정의하라. 단 if 문을 사용해서 수를 비교하세요.   
max(10, 2, 5)

[출력결과]  
10

```python
def print_max(a, b, c):
    maxValue = 0
    if a > b:
        maxValue = a
    else:
        maxValue = b
    
    if (c > maxValue):
        maxValue = c
    
    return maxValue

print(print_max(5, 10, 2))
```

[문제5]    
2 와 5의 배수가 아닌 갯수와 합계를 구하는 함수를 구현하세요.    

resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)  

[출력결과]    
2 와 5의 배수가 아닌 갯수 : 7    
2 와 5의 배수가 아닌 합계 : 107   

==============================================  
result(5,2,7,9,10,8,6)     

[출력결과]    
2 와 5의 배수가 아닌 갯수 :  2
2 와 5의 배수가 아닌 합계 :  16

```python
def proc(*t):
    newList = list(filter(lambda x: x % 2 != 0 and x % 5 != 0, t))
    return len(newList), sum(newList)

# count, total = proc(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)
count, total = proc(5, 2, 7, 9, 10, 8, 6)
print(count, total)
```
