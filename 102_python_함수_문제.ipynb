[문제1]  
문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 정의하세요.    
make_url("naver")  

[출력결과]    
www.naver.com  


```python
def make_url(url):
    newUrl = 'www.' + url + '.com'
    return newUrl

print(make_url('naver'))
```

문제6~13) 아래 함수를 호출했을때 정상적으로 실행이 되도록 호출하는 함수를 구현하세요.

```python
# 문제6)
def calc(a, b, /, c, *, d):
    return a + b + c + d
```

```python
# 문제7)
def introduce(name, /, age, *, country):
    return f"{name}, {age} years old from {country}"
```

```python
# 문제8)
def f(x, *args, y, **kwargs):
    return x, args, y, kwargs
```

```python
# 문제9)
def display(*args, **kwargs):
    print("ARGS:", args)
    print("KWARGS:", kwargs)
```

```python
# 문제10)
def config(*, debug=False, verbose=False):
    return f"Debug={debug}, Verbose={verbose}"
```

```python
# 문제11)
def info(name, /, *, age, country):
    return f"{name}, {age}, {country}"
```

```python
# 문제12)
def combine(a, /, b, *, c):
    return a + b + c
```

```python
#문제 13)
def test(a, b=2, *args, c=3, **kwargs):
    return a, b, args, c, kwargs
```
