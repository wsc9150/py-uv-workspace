문제1] 이름과 년도을 출력하도록 구현

['(이동민, 2019)', '(최재영, 2019)', '(Lion, 2018)']

```python
example1 = '이동민 교수님은 다음과 같이 설명했습니다(이동민,   2019). 그런데 다른 교수님은 이 문제에 대해서 다른 견해를 가지고 있었습니다(최재영, 2019). 또 다른 견해도 있었습니다(Lion, 2018)'
```

```python
pattern = r'\([a-zA-Z가-힣]+,\s+[0-9]+\)'
regex = re.compile(pattern)
result = regex.findall(example1)
result = [ i.replace(' ', '').replace(',', ', ') for i in result ]
print(result)
```

문제9] 등장인물 모으기

['All', 'Ross', 'Monica', 'Rachel', 'Customer', 'Chandler', 'Note', 'Scene', 'Paul', 'Phoebe', 'Waitress', 'Joey', 'Frannie']

```python
pattern = '^[a-zA-Z\(\[]+(?=:)'
regex = re.compile(pattern, re.M)
result = regex.findall(friends)
result = list(set(result))
result = [ re.sub('(\[|\()', '', i) for i in result ]
print(result)
```

문제13] data에 저장된 데이터에서 email만 추출


['greatking@naver.com', 'travel@daum.net', 'apple@gmail.com', 'life@abc.co.kr']


```python
data = '제 이메일 주소는 greatking@naver.com입니다. 오늘 저는 travel@daum.net 라는 주소로 메일을 보내려고 합니다. 저는 apple@gmail.com, life@abc.co.kr 라는 메일도 사용하고 있습니다.'
```

문제14] 숫자에서  천단위를 제거

문제15] 아이디 정규 표현식
   <조건>       
   1. 전체 길이 5~8자   
   2. 영문자로 시작     
   3. 숫자가 2개이상 포함되여 있어야함  
   4. 2자이상의 반복문자가 연속 2번 있으면 안됨  

```python
user_id = input()

if len(user_id) < 5 or len(user_id) > 8:
    print('입력한 아이디 길이가 맞지 않습니다. 다시 입력해주세요.')

if not re.match('^[A-Za-z]', user_id):
    print('아이디가 영문자로 시작하지 않습니다. 다시 입력해주세요.')

if len(''.join(re.findall('[0-9]', user_id))) < 2:
    print('아이디에 숫자가 최소 2개 이상 필요합니다. 다시 입력해주세요.')

if re.findall(r'(.{2,})\1', user_id):
    print('2자 이상의 반복 문자가 연속 2번으로 들어가있습니다 다시 입력해주세요.')
```
