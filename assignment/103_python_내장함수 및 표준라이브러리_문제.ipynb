문제 1) 무한 제너레이터에서 조건 만족하는 첫 N개 추출<br/>
1부터 시작하는 자연수 중 3의 배수만 필터링한 뒤, 처음 10개를 출력하세요. 

[출력]  
3, 6, 9, ..., 30

````python
import math

def infinite_generator(n):
    i = 1

    while i <= n:
        yield i
        i += 1

result = []
generator = infinite_generator(math.inf)

for i in generator:
    if i % 3 == 0:
        result.append(i)
    
    if (len(result) == 10):
        break

print(result)
````

문제 2) 한 줄에서 단어 빈도수 계산<br/>
파일의 한 줄에 있는 단어들의 빈도수를 계산하고, 빈도순으로 출력하세요.  

```
# 예: line = "data science is data"
# 출력: [('data', 2), ('science', 1), ('is', 1)]
```

````python
line = 'data science is data'
split_list = line.split()

word_dict = dict()
for word in split_list:
    if word not in word_dict.keys():
        word_dict[word] = 1
    else:
        word_dict[word] += 1

result = []
for key in word_dict.keys():
    result.append((key, word_dict[key]))

print(result)
````

문제 3) map과 lambda로 함수형 처리<br/>
리스트의 모든 실수 값에 자연로그를 취한 후 소수점 3자리까지 반올림하여 출력하세요.  
입력: [1.0, 2.718, 7.389]
출력: [0.0, 1.0, 2.0]  

````python
import math

value_list = [1.0, 2.718, 7.389]
# result = [ round(math.log(value), 3) for value in value_list ]
result = list(map(lambda x: round(math.log(x), 3), value_list))
print(result)
````

문제 4) datetime을 이용한 주간 날짜 생성<br/>
2025년 1월 1일부터 시작해 매주 수요일 날짜 10개를 리스트로 출력하세요.

문제 5) sorted의 key 인자 활용<br/>
다음 리스트에서 길이순 정렬 후, 알파벳 역순으로 2차 정렬하세요.  
["banana", "apple", "kiwi", "avocado", "fig"]

[출력]  
['fig', 'kiwi', 'apple', 'banana', 'avocado']

````python
fruit_list = ["banana", "apple", "kiwi", "avocado", "fig"]
result = sorted(fruit_list, key=lambda x: (-len(x), x), reverse=True)
print(result)
````

문제 6) groupby로 구간 그룹핑<br/>
[75, 82, 90, 66, 72, 88, 55] 점수를 10점 단위로 그룹핑해서 출력하세요.    
```
{
  '50~59': [55],
  '60~69': [66],
  ...
}
```

출력 : {'70~79': [75, 72], '80~89': [82, 88], '90~99': [90], '60~69': [66], '50~59': [55]}

````python
from itertools import groupby

def group_key(n):
    return n // 10

score_list = [75, 82, 90, 66, 72, 88, 55]
score_list.sort()

grouped = groupby(score_list, key=lambda x: x // 10)

result = dict()
for key, group in grouped:
    result[str(key * 10) + '~' + str(key * 10 + 9)] = list(group)

print(result)
````
