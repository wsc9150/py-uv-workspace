```
[문제1]
분자(a)와 분모(b)를 받아 나눗셈 연산을 하되, 분모가 0일 때 프로그램이 정지하지 않고 "0으로 나눌 수 없습니다."를 반환하는 proc_divide 함수를 구현하세요.

[출력결과]
5.0
0으로 나눌 수 없습니다.
```

````python
def proc_divide(a, b):
    try:
        result = a / b
        print(result)
    except ZeroDivisionError as e:
        print('0으로 나눌 수 없습니다.')

# proc_divide(5, 1)
proc_divide(5, 0)
````

```
[문제2]
매개변수로 지정한 텍스트 파일 경로를 조회하여 읽어오되, 해당 파일이 디렉토리에 없어도 예외 크래시 없이 "파일이 존재하지 않습니다."를 출력하는 proc_safe_open 함수를 구현하세요.

[출력결과]
파일이 존재하지 않습니다.
```

````python
def proc_safe_open(path):
    try:
        fread = open(path, 'r')
        result = fread.read()
        print(result)
    except FileNotFoundError as e:
        print('파일이 존재하지 않습니다.')

# proc_safe_open('../source/test.txt')
proc_safe_open('../source/text.txt')
````

```
[문제5]
리스트 내의 문자열 데이터 숫자들을 모두 더하는 과정에서 문자가 섞여 연산 오류(TypeError)가 발생하면, 무시하고 숫자 형태의 데이터만 합산하여 결과를 반환하는 proc_sum_list 함수를 구현하세요.

[출력결과]
60
```

````python
def proc_sum_list(value_list):
    try:
        print(sum(value_list))
    except TypeError as e:
        num_list = [ i for i in value_list if type(i) == int ]
        print(sum(num_list))

# proc_sum_list([1, 2, 3, 4, 5])
proc_sum_list([1, 2, 3, None, 'a', True])
````

```
[문제6]
사용자로부터 양수만 입력받아야 하는 함수에서, 만약 음수(0 미만)가 입력값으로 들어오면 raise 키워드를 사용하여 ValueError를 강제로 발생시키고 에러 메시지로 "음수는 허용되지 않습니다."를 송출하는 proc_check_positive 함수를 구현하세요.

[출력결과]
```

````python
def proc_check_positive(value):
    try:
        if value <= 0:
            raise ValueError
        print(value)
    except ValueError as e:
        print('음수는 허용되지 않습니다.')

# proc_check_positive(5)
proc_check_positive(-3)
````

```
[문제8]
하나의 try 블록 안에서 두 가지 이상의 에러가 발생할 여지가 있을 때, ValueError와 ZeroDivisionError를 멀티 except 구문으로 연쇄 캐칭하여 각각 다른 에러 안내 메시지를 출력하는 proc_multi_exception 함수를 구현하세요.

[출력결과]
에러: 숫자로 변환할 수 없는 입력입니다.
에러: 0으로 나눌 수 없습니다.
```

````python
def proc_multi_exception(a, b):
    try:
        num_a = int(a)
        num_b = int(b)

        print(num_a / num_b)
    except ValueError as e:
        print('에러 : 숫자로 변환할 수 없는 입력입니다.')
    except ZeroDivisionError as e:
        print('에러 : 0으로 나눌 수 없습니다.')

# proc_multi_exception('12', '4')
proc_multi_exception('a', '4')
proc_multi_exception('12', '0')
````

```
[문제9]
문자열 형식의 숫자를 받아 나눗셈 연산을 수행하되, 예외가 발생하지 않았을 때만 작동하는 else 블록을 결합하여 "연산에 정상 성공했습니다." 문구와 결과값을 출력하는 proc_success_check 함수를 구현하세요.

[출력결과]
연산에 정상 성공했습니다. 결과: 10.0
```

````python
def proc_success_check(a, b):
    try:
        num_a = int(a)
        num_b = int(b)
        result = num_a / num_b
    except Exception as e:
        print(e)
    else:
        print('연산에 정상 성공했습니다. 결과 :', result)

proc_success_check('6', '3')
proc_success_check('6', 'a')
````

```
[문제10]
try-except 구조로 파일 처리 기능을 수행하면서, 작업 도중 예외 에러가 뜨거나 안 뜨거나 상관없이 언제나 함수 탈출 직전에 실행되어 확실하게 시스템 자원을 반환(close)해 주는 finally 블록을 포함한 proc_final_close 함수를 구현하세요.

[출력결과]
예외 핸들링: 파일을 찾지 못해 안전 모드로 전환합니다.
finally 블록: 파일 스트림 자원 체크를 완수합니다.
```

````python
def proc_final_close(path):
    fread = None

    try:
        fread = open(path, 'r')
        fread.read()
        fread.close()
    except Exception as e:
        print('예외 핸들링 : 파일을 찾지 못해 안전 모드로 전환합니다.')
    finally:
        if fread is not None:
            fread.close()
        print('finally 블록 : 파일 스트림 자원 체크를 완수합니다.')

proc_final_close('../source/test.txt')
proc_final_close('../source/text.txt')
````

```

파이썬 딕셔너리 데이터를 입력받아 직렬화된 JSON 형태의 문자열(String) 객체로 변환하여 반환하는 proc_to_json 함수를 구현하세요.

[출력결과]
'{"name": "홍길동", "age": 20, "city": "서울"}'
```

````python
import json

def proc_to_json(dict_obj):
    result = json.dumps(dict_obj, ensure_ascii=False)
    return result

print(proc_to_json({ 'name': '홍길동', 'age': 20, 'city': '서울' }))
````
